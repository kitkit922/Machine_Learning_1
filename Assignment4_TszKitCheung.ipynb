# Evaluate clustering performance using CH-Index and Silhouette score. (For Epsilon 1 - 20 and min samples 5)


```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score

CHlist = []
SSlist = []

max_ch = 0
max_ss = 0
max_ch_n = 0
max_ss_n = 0


for i in range (1,51):
  dbscan = DBSCAN(eps=i, min_samples = 5)
  y_dbscan = dbscan.fit_predict(Xm)

  # calinski_harabasz_score
  try:
    chscore = calinski_harabasz_score(Xm,y_dbscan)
    CHlist.append(chscore)
    print(f'Calinski Harabasz Score(eps = {i}) : {chscore}')

    # silhouette_score
    ssscore = silhouette_score(Xm,y_dbscan)
    SSlist.append(ssscore)
    print(f'Silhouette Score(eps = {i}) : {ssscore}')
    print('')
  except:
    pass

#find Calinski Harabasz Score max
for i, element in enumerate(CHlist):
  if element > max_ch:
    max_ch = element
    max_ch_n = i
print(f'Max Calinski Harabasz Score: (eps = {max_ch_n + 2}) : {max_ch}')

#find Silhouette Score max
for i, element in enumerate(SSlist):
  if element > max_ss:
    max_ss = element
    max_ss_n = i
print(f'Max Silhouette Score: (eps = {max_ss_n + 2}) : {max_ss}')


```

Max Calinski Harabasz Score when epsilon (eps = 10)     
Max Silhouette Score when epsilon (eps = 9)

#Compare DBSCAN clustering with K-means
I use 3 situations  
DBSCAN (For Epsilon 5 and min samples 5), cluster = 5     
DBSCAN (For Epsilon 10 and min samples 5), cluster = 4      
DBSCAN (For Epsilon 10 and min samples 12), cluster = 3     

```python
from sklearn.cluster import DBSCAN, KMeans
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
dbscan = DBSCAN(eps=5, min_samples = 5)
y_kmeans = kmeans.fit_predict(Xm)
y_dbscan = dbscan.fit_predict(Xm)

fig,ax = plt.subplots(1,2,figsize=(12,6))
colorArray = [
    (0.2, 0.4, 0.6),
    (0.8, 0.1, 0.3),
    (0.5, 0.7, 0.2),
    (0.9, 0.5, 0.1),
    (0.3, 0.6, 0.9),
    (0.7, 0.2, 0.8),
    (0.1, 0.9, 0.4),
    (0.4, 0.8, 0.7),
    (0.6, 0.3, 0.5),
    (0.2, 0.5, 0.8)
]

for i in range(5):
  ax[0].scatter(Xm[y_kmeans == i, 0], Xm[y_kmeans == i, 1], s = 100, c = colorArray[i], label = f'Cluster {i+1}')
ax[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 150, c = 'yellow', label = 'Centroids')
ax[0].set_title('Clusters (n=5)', fontsize = 25)
ax[0].set_xlabel('Annual Income (k$)', fontsize = 15)
ax[0].set_ylabel('Spending Score (1-100)', fontsize = 15)
ax[0].legend(fontsize = 15)

sns.scatterplot(data = Xm, x = Xm[:,0], y = Xm[:,1], hue = y_dbscan, palette = 'Set1')
ax[1].set_title('Clusters (Epsilon 5 and min samples 5)', fontsize = 25)
ax[1].set_xlabel('Annual Income (k$)', fontsize = 15)
ax[1].set_ylabel('Spending Score (1-100)', fontsize = 15)
ax[1].legend(fontsize = 15)
plt.show()

# calinski_harabasz_score for kmeans
chscore_km = calinski_harabasz_score(Xm,y_kmeans)
print(f'Calinski Harabasz Score for K-means: {chscore_km}')

# silhouette_score for kmeans
ssscore_km = silhouette_score(Xm,y_kmeans)
print(f'Silhouette Score for K-means: {ssscore_km}')
print('')

# calinski_harabasz_score for dbscan
chscore_db = calinski_harabasz_score(Xm,y_dbscan)
print(f'Calinski Harabasz Score for DBSCAN: {chscore_db}')

# silhouette_score for dbscan
ssscore_db = silhouette_score(Xm,y_dbscan)
print(f'Silhouette Score for DBSCAN: {ssscore_db}')
print('')

if chscore_km > chscore_db:
  print('K-means has a higher calinski harabasz score')
else:
  print('DBSCAN has a higher calinski harabasz score')

if ssscore_km > ssscore_db:
  print('K-means has a higher Silhouette Score')
else:
  print('DBSCAN has a higher Silhouette Score')
```
